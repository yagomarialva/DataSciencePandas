## Seleções e Frequencias

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/database_norte.csv', sep = ';')
```

```python
dados
```

```python
# Selecione somente os casos classificados com tipo 'AM'.
selecao = dados['estado'] == 'AM'
n1 = dados[selecao].shape[0]
n1
```

```python
# Selecione as cidades classificados com tipos 'Manaus', 'Parintins' e 'Iranduba'.
selecao = (dados['municipio'] == 'AM') | (dados['municipio'] == 'Manaus') | (dados['municipio'] == 'Parintins') | (dados['municipio'] == 'Iranduba') 
n2 = dados[selecao].shape[0]
n2
```

```python
# Selecione os cidades entre 100 e 300 casos diarios confirmados, incluindo os limites.
selecao = (dados['casosNovos'] >= 100) & (dados['casosNovos'] <= 300) 
n3 = dados[selecao].shape[0]
n3
```

```python
# Selecione os cidades que tenham pelo menos casos acumulados menor que 1.000,00.
```

```python
selecao = (dados['casosAcumulado'] >= 10000) & (dados['estado'] == 'AM')
n4 = dados[selecao].shape[0]
n4
```
